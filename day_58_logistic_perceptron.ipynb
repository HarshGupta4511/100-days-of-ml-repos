```python
df = pd.read_csv('/kaggle/input/datasets/johnsmith88/heart-disease-dataset/heart.csv')
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pred = clf1.predict(X_test)
y_pred1 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print('Acc_LR',accuracy_score(y_test,y_pred))
print('Acc_DT',accuracy_score(y_test,y_pred1))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred),columns=list(range(0,2)))
```

```python
print("DT Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pred
result['Decision Tree Prediction'] = y_pred1
```

```python
result.sample(10)
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
print('Logistic Reg')
print('-'*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred),columns=list(range(0,2)))
print(cdf)
print('-'*50)
print('Precision',precision_score(y_test,y_pred))
print('recall_score',recall_score(y_test,y_pred))
print('f1_score',f1_score(y_test,y_pred))
```

```python
print('DT')
print('-'*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
print(cdf)
print('-'*50)
print('Precision',precision_score(y_test,y_pred1))
print('recall_score',recall_score(y_test,y_pred1))
print('f1_score',f1_score(y_test,y_pred1))
```
