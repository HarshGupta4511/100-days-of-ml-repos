```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

from sklearn.decomposition import PCA
pca = PCA(n_components=None)
x_train_trf = pca.fit_transform(x_train)
x_test_trf = pca.transform(x_test)

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train_trf,y_train)

y_pred = lr.predict(x_test_trf)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

from sklearn.decomposition import PCA
pca = PCA(n_components=8)
x_train_trf = pca.fit_transform(x_train)
x_test_trf = pca.transform(x_test)

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train_trf,y_train)

y_pred = lr.predict(x_test_trf)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

The original dataset contained 30 numerical features. PCA was applied after feature standardization, and the number of principal components was selected to retain approximately 95% of the total variance. This reduced the feature space from 30 to 8 dimensions. Logistic Regression trained on the PCA-transformed data achieved an accuracy of 99%, compared to 97% on the original feature set. The result suggests that PCA effectively removed redundant information while preserving the most informative patterns for classification.

```python
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Selecting the Number of Principal Components")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

train_pred = lr.predict(x_train_trf)
test_pred = lr.predict(x_test_trf)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy :", accuracy_score(y_test, test_pred))
```
